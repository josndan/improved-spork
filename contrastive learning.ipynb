```python
from pl_bolts.models.self_supervised import SimCLR
from pl_bolts.models.self_supervised.simclr import SimCLREvalDataTransform, SimCLRTrainDataTransform
import torchvision.datasets as dset
import torchvision.transforms as T
from pytorch_lightning import LightningModule, Trainer
from torch.utils.data import DataLoader
```

```python
transform = T.Compose([
                # T.ToTensor(),
                SimCLRTrainDataTransform(),
                # T.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))
            ])

cifar10_train = dset.CIFAR10('./Datasets', train=True, download=True,
                             transform=transform)
cifar10_val = dset.CIFAR10('./Datasets', train=True, download=True,
                           transform=transform)
```

```python
train_dataset = cifar10_train
val_dataset = cifar10_val
```

```python
model = SimCLR(gpus=1,num_samples=len(train_dataset),batch_size=64,dataset="")
```

```python
trainer = Trainer(max_epochs=10)
trainer.fit(
    model,
    DataLoader(train_dataset,batch_size=32,drop_last=True),
    DataLoader(val_dataset,batch_size=32,drop_last=True),
)
```
